```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn import tree
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sklearn.datasets.load_iris()
model = tree.DecisionTreeClassifier()
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state = 100)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model,X_train,y_train,cv = 5,scoring='accuracy')
print(scores)
```

```python
print("average = ", np.average(scores))
print("standard deviation = ",np.std(scores))
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(classification_report(y_test,model.predict(X_test)))
sns.heatmap(confusion_matrix(y_test,model.predict(X_test)), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
print("before pruning : ")
tree.plot_tree(model)
```

```python
model = tree.DecisionTreeClassifier(max_depth=3)
model.fit(X_train,y_train)
tree.plot_tree(model)
```

```python
model = tree.DecisionTreeClassifier(min_samples_leaf=4)
model.fit(X_train,y_train)
tree.plot_tree(model)
```

```python
model = tree.DecisionTreeClassifier(min_samples_split=100)
model.fit(X_train,y_train)
tree.plot_tree(model)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = tree.DecisionTreeClassifier()
clf = GridSearchCV(model,param_grid={
    "max_depth" : [2,3,4],
    "min_samples_split"  : [100,120,150],
    "min_samples_leaf" : [3,4]
},scoring="accuracy",cv=3)
clf.fit(X_train,y_train)
```

```python
print(clf.best_params_)
```
